```python
neg='/content/gdrive/MyDrive/Dataset/Negative'
data_n=os.listdir(neg)
pos='/content/gdrive/MyDrive/Dataset/Positive'
data_p=os.listdir(pos)
```

```python
# for f in data_n:
#   print(f)
#   break
```

```python


import numpy as np
from PIL import Image 
x=[]
y=[]

for im  in data_n:
    im_gray=np.array(Image.open(f'/content/gdrive/MyDrive/Dataset/Negative/{im}').convert('L'))
    thresh= 170
    maxval= 255
    im_bin= (im_gray > thresh)*maxval
    Image.fromarray(np.uint8(im_bin)).save(f'bin{im}.jpg')
    
    x.append(im_bin.flatten())
    y.append(0)
    
for im in data_p:
    im_gray=np.array(Image.open(f'/content/gdrive/MyDrive/Dataset/Positive/{im}').convert('L'))
    thresh= 120
    maxval= 255
    im_bin= (im_gray > thresh)*maxval
    Image.fromarray(np.uint8(im_bin)).save(f'bin{im}.jpg')
    
    x.append(im_bin.flatten())
    y.append(1)

print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test, y_train, y_test=train_test_split(x,y,test_size=0.3)
print(y_test)
```

```python
from sklearn.svm import SVC
sv=SVC(kernel='sigmoid')
sv.fit(x_train,y_train)
y_pred=sv.predict(x_test)
# from sklearn.linear_model import LogisticRegression 
# model=LogisticRegression()
# model.fit(x_train,y_train,sample_weight=None)
# y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score
print(accuracy_score(y_test, y_pred))
print(precision_score(y_test,y_pred))
```
